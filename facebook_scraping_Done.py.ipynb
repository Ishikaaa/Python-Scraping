```python
# access page of post 
z=[]
a_p_p=0
for i in soup_profile_page.find_all('a',href=True):
    #time.sleep(4)
    if i['href'][0:11]=="/story.php?" :
        a_p_p+=1
        if a_p_p==0 or a_p_p%3==0:
            z.append((i['href']))
print(z)
```

```python
# access profile post page 
a_p_p_p=[]
for y in z:
    url_p_p_p="https://m.facebook.com/"+y
    data_p_p_p=requests.get(url_p_p_p,cookies=cj)
    soup_p_p_p=bs4.BeautifulSoup(data_p_p_p.text,'html.parser')
    for zzz in soup_p_p_p.title:
        if zzz.lower()=="comments":
            zzz="None"
        a_p_p_p.append(zzz)
print(a_p_p_p)        
```
