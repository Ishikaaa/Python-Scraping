```python
import requests 
from bs4 import BeautifulSoup
print("Modules imported successfully")
```

```python
tweet_id=[]
tweet_text=[]
```

```python
url="https://twitter.com/TheOnion"
data=requests.get(url)
htmll=BeautifulSoup(data.text,'html.parser')
print(htmll)
```

```python
timeline=htmll.select('#timeline li.stream-item')
print("Done")
```

```python
print(timeline)
```

```python
for tweet in timeline:
    tweet_id.append(tweet['data-item-id'])
    tweet_textt=tweet.select('p.tweet-text')[0].get_text()
    tweet_text.append(tweet_textt)
print(tweet_id)
print(tweet_text)
```

```python
import pandas as pd 
aa={"id":tweet_id,"tweet":tweet_text}
df=pd.DataFrame(aa)
print(df)
```

```python
df.to_csv("Twitter.csv")
print("done")
```
